```python
#Letâ€™s get the total crimes committed per state over the years:

# Group by state and get total crimes
statewise_total = df.groupby('STATE_UT')['TOTAL_IPC_CRIMES'].sum().sort_values(ascending=False)

# Display top 10
statewise_total.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
# Total crimes per year
yearly_crimes = df.groupby('YEAR')['TOTAL_IPC_CRIMES'].sum()

# Plot
plt.figure(figsize=(10,6))
plt.plot(yearly_crimes.index, yearly_crimes.values, marker='o', color='blue')
plt.title('Total IPC Crimes in India (2001â€“2012)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Total IPC Crimes')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Count missing values
print(df.isnull().sum())

```

```python
# Visual heatmap of nulls
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
#Crimes per Year:
df['YEAR'].value_counts().sort_index().plot(kind='bar', figsize=(10,4))
plt.title("Entries per Year")
plt.xlabel("Year")
plt.ylabel("Number of Districts")
plt.grid(True)
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(15, 12))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='coolwarm')
plt.title("Correlation Between Crime Categories")
plt.show()
```

```python
# Top 10 states by total IPC crimes
top_states = df.groupby('STATE/UT')['TOTAL IPC CRIMES'].sum().sort_values(ascending=False).head(10)
top_states.plot(kind='bar', color='crimson')
plt.title("Top 10 States with Highest Total IPC Crimes")
plt.ylabel("Total Crimes")
plt.xticks(rotation=45)
plt.show()
```

```python
print(df.columns.tolist())
```

```python
# Yearly total crimes across India
df.groupby('YEAR')['TOTAL IPC CRIMES'].sum().plot(kind='line', marker='o')
plt.title("Crime Trend Over Time in India")
plt.xlabel("Year")
plt.ylabel("Total IPC Crimes")
plt.grid(True)
plt.show()
```

```python
df['MURDER'].hist(bins=30, color='orange', edgecolor='black', figsize=(8, 5))
plt.title("Distribution of Murder Cases Across Districts")
plt.xlabel("Number of Murder Cases")
plt.ylabel("Number of Districts")
plt.grid(False)
plt.show()
```

```python
df.groupby('YEAR')['MURDER'].sum().plot(label='Murder')
df.groupby('YEAR')['RAPE'].sum().plot(label='Rape')
df.groupby('YEAR')['KIDNAPPING & ABDUCTION'].sum().plot(label='Kidnapping')
plt.legend()
plt.title("Crime Trends Over Time")
plt.show()
```
